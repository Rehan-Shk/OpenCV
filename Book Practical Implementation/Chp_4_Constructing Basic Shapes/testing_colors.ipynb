```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
colors = {'blue': (255, 0, 0), 'green': (0, 255, 0), 'red': (0, 0, 255),
'yellow': (0, 255, 255), 'magenta': (255, 0, 255), 'cyan': (255, 255, 0),
'white': (255, 255, 255), 'black': (0, 0, 0), 'gray': (125, 125, 125),
'rand': np.random.randint(0, high=256, size=(3,)).tolist(), 'dark_gray':
(50, 50, 50), 'light_gray': (220, 220, 220)}

colors_names = list(colors.keys())
print(colors_names)

def show_with_matplotlib(image, title):
    image = image[:, :, ::-1]

    fig, ax = plt.subplots()

    # Set Y-axis ticks and labels to color names
    yticks = np.arange(40, 40 * len(colors) + 1, 40)
    ax.set_yticks(yticks)
    ax.set_yticklabels(colors_names)
    
    # Hide the X-axis
    ax.get_xaxis().set_visible(False)
    
    # Display the image using matplotlib with Y-axis labels
    ax.imshow(image)  # Convert BGR to RGB for plt display
    plt.title('Testing Colors')
    plt.show()
```

```python
image = np.zeros((500, 500, 3), dtype='uint8')
image[:] = colors['light_gray']
seperation = 40

for key in colors_names:
    cv2.line(image, (0, seperation), (500, seperation), colors[key], 10)

    seperation += 40

show_with_matplotlib(image, 'Testing Colors')
```
