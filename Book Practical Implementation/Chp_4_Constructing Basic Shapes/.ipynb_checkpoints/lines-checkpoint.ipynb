```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
colors = {'blue': (255, 0, 0), 'green': (0, 255, 0), 'red': (0, 0, 255),
'yellow': (0, 255, 255), 'magenta': (255, 0, 255), 'cyan': (255, 255, 0),
'white': (255, 255, 255), 'black': (0, 0, 0), 'gray': (125, 125, 125),
'rand': np.random.randint(0, high=256, size=(3,)).tolist(), 'dark_gray':
(50, 50, 50), 'light_gray': (220, 220, 220)}
```

```python
def show_with_matplotlib(image, title):
    image = image[: ,: ,::-1]
    plt.imshow(image)
    plt.title(title)
    plt.show()
```

```python
image = np.zeros((400, 400, 3), dtype="uint8")
image[:] = colors['light_gray']
show_with_matplotlib(image, 'Figure 1')
```

```python
cv2.line(image, (0, 0), (400, 400), colors['green'], 20, lineType=8)
cv2.line(image, (0, 400), (400, 0), colors['black'], 20, lineType=4)
cv2.line(image, (0, 200), (400, 200), colors['yellow'], 10, lineType=cv2.LINE_AA)
cv2.line(image, (400, 400), (400,800), colors['red'], 10, lineType=cv2.LINE_AA, shift=1)

show_with_matplotlib(image, 'Lines')
```
