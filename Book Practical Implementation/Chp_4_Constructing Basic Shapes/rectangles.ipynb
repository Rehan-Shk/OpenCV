```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_with_matplotlib(image, title):
    image = image[:, :, ::-1]
    plt.imshow(image)
    plt.title(title)
    plt.show()
```

```python
colors = {'blue': (255, 0, 0), 'green': (0, 255, 0), 'red': (0, 0, 255),
'yellow': (0, 255, 255), 'magenta': (255, 0, 255), 'cyan': (255, 255, 0),
'white': (255, 255, 255), 'black': (0, 0, 0), 'gray': (125, 125, 125),
'rand': np.random.randint(0, high=256, size=(3,)).tolist(), 'dark_gray':
(50, 50, 50), 'light_gray': (220, 220, 220)}
```

```python
image = np.zeros((400, 400, 3), dtype='uint8')

cv2.rectangle(image, (10, 50), (60, 350), colors['green'], 3, lineType=4)
cv2.rectangle(image, (80, 50), (130, 350), colors['cyan'], -1, lineType=cv2.LINE_AA)
cv2.rectangle(image, (140, 50), (390, 190), colors['yellow'], 3, lineType=8)
cv2.rectangle(image, (140, 210), (390, 350), colors['rand'], -1, lineType=4)

show_with_matplotlib(image, 'Rectangles')
```
