```python
import cv2
```

```python
capture = cv2.VideoCapture('webcam.mp4')
```

```python
last_frame_index = capture.get(cv2.CAP_PROP_FRAME_COUNT) - 1
print(f'Last Index : {last_frame_index}')

while capture.isOpened() and last_frame_index > 0:
    capture.set(cv2.CAP_PROP_POS_FRAMES, last_frame_index)
    
    ret, frame = capture.read()
    if ret:
        cv2.imshow('Frame', frame)
        last_frame_index -= 1
        print(f'Current Index : {last_frame_index}')

        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
    else:
        break

capture.release()
cv2.destroyAllWindows()
```
